```
from math import tan, pi
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import time


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


def my_tan( x , epsilon):
  signature_flag = 1
  mic = 10**(-12)
  d=0
  j=1
  a=x
  c = f = mic 
  if x < 0:
    x *= -1 
    signature_flag = -1
    while True:
      b = 2*(j-1)+1 
      d = max( b + a *d , mic)
      c = max(b + a/c ,mic)
      a = -x ** 2
      d = 1/d
      delta = c* d
      f = delta * f
      j += 1
      if abs(delta -1) <= epsilon:
        break
    
  return signature_flag*f

if __name__ == "__main__":
  labels = []
  men_means = []  
  women_means = [] 
  start =time.time()
  for _ in range(10000):
    x = random.uniform(-pi/2,pi/2)
    labels.append(x)
    men_means.append(my_tan(x,1000))
    women_means.append(tan(x))
    sum = abs(tan(x)-my_tan(x,1000))

  stop = time.time()
  consumed_time = stop - start
  err = sum /10000

  x = np.arange(len(labels))  # the label locations
  width = 0.25  # the width of the bars

  fig, ax = plt.subplots()
  rects1 = ax.bar(x - width/2, men_means, width, label='my_tan')
  rects2 = ax.bar(x + width/2, women_means, width, label='tan')

  # Add some text for labels, title and custom x-axis tick labels, etc.
  ax.set_title('Diferente valori mytan tan')
  ax.set_xticks(x)
  ax.set_xticklabels(labels)
  ax.legend()



  autolabel(rects1)
  autolabel(rects2)

  fig.tight_layout()
  plt.show()

  my_tan(10,1000)
```
